```python
import numpy as np
import matplotlib.pyplot as plt
```

> 使用Numpy的arange方法生成了[0, 0.1, 0.2, ... , 5.9]的数据，然后对其各个元素调用NumPy的sin函数。

```python
#生成数据
x=np.arange(0,6,0.1) #以0.1为单位步长，生成0到6的数据
y1=np.sin(x)
y2=np.cos(x)

#绘制图形
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="-.")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin and cos')
plt.legend()
plt.show()
```
